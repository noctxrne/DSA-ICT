```python
import numpy as np
import pandas as pd
url="https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.csv"
data=pd.read_csv(url)
df=pd.DataFrame(data)
```

```python
df['place'].nunique()
```

```python
len(df)
```

```python
df.head()
```

```python
df['time'] = pd.to_datetime(df['time'])
```

```python
df.info()
df.isnull().sum()
```

```python
df = df.dropna(subset=['latitude', 'longitude', 'mag'])
df
df.info()
```

```python
df[df['mag']>=4]
```

```python
df['day_name'] = df['time'].dt.day_name()
df.head()
```

```python
df['severity_level']=np.where(df['mag'] < 4, 'Low',
                     np.where(df['mag'] > 6.0, 'High', 'Moderate'))
df
```

```python
df['place'].value_counts()
```

```python
earthquake_summary = df.groupby(df['time'].dt.date).agg({'depth':'max', 'mag':'mean'}).reset_index()
earthquake_summary.columns = ['date', 'max_depth', 'average_magnitude']
earthquake_summary
```

```python
df.to_csv('cleaned_earthquakes.csv')
df
```

```python
earthquake_summary.to_csv('earthquake_summary.csv')
earthquake_summary
```

```python
import sqlite3
conn=sqlite3.connect("earthquake_summary.db")
```
