```python
import pandas as pd
population=pd.read_csv("/content/population.csv")
gdp=pd.read_excel("/content/gdp.xlsx")
internetUsers=pd.read_json("/content/internet_users.json")
literacyRate=pd.read_xml("/content/literacy_rate.xml")
```

```python
population.rename(columns={'country': 'Country'}, inplace=True)
internetUsers.rename(columns={'country': 'Country'}, inplace=True)
literacyRate.rename(columns={'name': 'Country'}, inplace=True)

#inner:
merged_df = population.merge(gdp, on='Country', how='inner')\
              .merge(internetUsers, on='Country', how='inner')\
              .merge(literacyRate, on='Country', how='inner')
merged_df["internet penetration rate"]=merged_df["internet_users"]/merged_df["population"]*100
merged_df
```

```python
merged_df.sort_values(by="internet penetration rate",ascending=False).head()
```

```python
averageLiteracy=merged_df["literacy_rate"].mean()
averageLiteracy
```

```python
merged_df[['GDP', 'population', 'literacy_rate', 'internet penetration rate']].corr()
```

```python
import sqlite3
conn=sqlite3.connect("FileDB.db") #creating a database
merged_df.to_sql('Global_Data', conn, index=False, if_exists='replace') #converting merged_df into a table named Global Data inside FileDB database with no Index
# merged_df.execute("select*from merged_df")
pd.read_sql_query("SELECT * FROM Global_Data", conn)
```

```python
# a. Countries with highest internet penetration

pd.read_sql_query("SELECT Country ,`internet penetration rate` \
                     FROM Global_Data \
                     ORDER BY `internet penetration rate` DESC limit 3",conn)
```

```python
# b. Average literacy rate

pd.read_sql_query("SELECT AVG(literacy_rate) AS avg_literacy_rate FROM Global_Data",conn)
```

```python
# c. Correlation between literacy rate and internet penetration
merged_df['literacy_rate'].corr(merged_df['internet penetration rate'])
```

```python
# d. GDP per capita analysis
pd.read_sql_query("SELECT Country, GDP, population, \
                    ROUND(CAST(GDP AS FLOAT) / population, 2) AS gdp_per_capita \
                    FROM Global_Data \
                    ORDER BY gdp_per_capita DESC",conn)
```

```python
# e. Countries with GDP per capita above $10,000
pd.read_sql_query("SELECT Country, ROUND(CAST(GDP AS FLOAT) / population, 2) AS gdp_per_capita \
                    FROM Global_Data \
                    WHERE gdp_per_capita > 10000",conn)

```

```python
# f. Total population covered
pd.read_sql_query("SELECT SUM(population) AS total_population FROM Global_Data",conn)
```

```python
# g. Countries with lowest literacy rates and their internet access
pd.read_sql_query("SELECT literacy_rate,internet_users \
                    FROM Global_Data \
                    ORDER BY literacy_rate ASC limit 5  ",conn)
```

```python
# h. Top 5 wealthiest countries by GDP and population size

pd.read_sql_query("SELECT Country, GDP, population \
                    FROM Global_Data \
                    ORDER BY GDP DESC LIMIT 5 ",conn)
```

```python
# i. Countries where internet penetration > 70%

pd.read_sql_query("SELECT Country, `internet penetration rate` \
                FROM Global_Data \
                WHERE `internet penetration rate` > 70", conn)
```

```python
# j. Average GDP per capita for countries with internet penetration above 50%

pd.read_sql_query("SELECT AVG(gdp_per_capita)FROM (SELECT ROUND(CAST(GDP AS FLOAT) / population, 2) AS gdp_per_capita \
                    FROM Global_Data \
                    WHERE `internet penetration rate` > 50)",conn)


```

```python
# k. Number of countries with literacy rate above 90%, and their avg internet penetration

pd.read_sql_query("SELECT Country FROM Global_Data WHERE literacy_rate>90  ",conn)
```
