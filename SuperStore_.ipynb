## **Plotting Graph**

### **Univariate**

```python
#numerical

#Pie Chart representing the sub_categories and their frequency of them being ordered
Category_counts = ssdf.groupby("Category")["Sales"].sum()

plt.pie(Category_counts, labels=Category_counts.index, autopct="%1.1f%%")
plt.title("Sales Distribution of Catgories")
plt.show()
```

```python
#categorical

plt.figure(figsize=(8,10))

sns.histplot(ssdf["Sub-Category"], bins=40, edgecolor="black")
plt.xlabel("Sub-Categories")
plt.ylabel("Frequency")
plt.title("Histogram of Sub-Categories")
plt.xticks(rotation=45)
plt.show()
```

### **Bivariate**

```python
#numerical vs numerical (sales vs profit)

plt.figure(figsize=(7,5))
sns.scatterplot(x="Sales", y="Profit", data=ssdf, hue="Category", alpha=0.7)

plt.title("Bivariate Analysis: Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.xticks(rotation= 45)
plt.yticks(rotation= 45)
plt.show()

#The scatterplot shows the graphical representation data between sales and profit.
# The color hue indicates the different individual categories present
```

```python
#profit by category

plt.figure(figsize=(6,4))
sns.boxplot(x="Category", y="Profit", data=ssdf)
plt.title("Bivariate Analysis: Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.xticks(rotation= 45)
plt.yticks(rotation = 45)
plt.show()

# RESULT Analysis:
# Office Supplies is relatively stable, but still has both profit and loss cases.
# Furniture seems to generate smaller profits but also smaller losses.
# Technology is highly volatile : large profits but also large losses.
```

```python
# category vs category (region vs segment)

plt.figure(figsize=(8,5))
sns.countplot(x="Region", hue="Segment", data=ssdf)
plt.title("Customer Segment by Region")
plt.xlabel("Region")
plt.ylabel("Segment Count")
plt.xticks(rotation =45)
plt.show()

# RESULT Analysis:
# Segments separated by regions across their count.
# All 4 region has been represented the 3 segments(in different colors) and their counts respectively.
```

### **Multivariate**

```python
#categorical vs numerical (Sub-Categories vs Profit)

plt.figure(figsize=(30,10))

plt.plot(ssdf['Sub-Category'].unique(), ssdf.groupby('Sub-Category')['Profit'].sum(),'o-',color='blue')
plt.plot(ssdf['Sub-Category'].unique(), ssdf.groupby('Sub-Category')['Sales'].sum(),'d-',color='orange')

# Horizontal lines for Profit (blue)
plt.axhline(y=(ssdf.groupby('Sub-Category')['Profit'].sum()).max(), color='green', linestyle='--', label='Profit Max')
plt.axhline(y=(ssdf.groupby('Sub-Category')['Profit'].sum()).min(), color='green', linestyle=':', label='Profit Min')
plt.axhline(y=(ssdf.groupby('Sub-Category')['Profit'].sum()).mean(), color='blue', linestyle='-.', label='Profit Avg')


# Horizontal lines for Sales (orange)
plt.axhline(y=(ssdf.groupby('Sub-Category')['Sales'].sum()).max(), color='red', linestyle='--', label='Sales Max')
plt.axhline(y=(ssdf.groupby('Sub-Category')['Sales'].sum()).min(), color='red', linestyle=':', label='Sales Min')
plt.axhline(y=(ssdf.groupby('Sub-Category')['Sales'].sum()).mean(), color='orange', linestyle='-.', label='Sales Avg')

plt.title('Sub-Categories vs Profit & Sales')
plt.xlabel('Sub-Categories')
plt.ylabel('Profit and Sales')
plt.legend(["Profit","Sales"])
plt.show()

# RESULT Analysis:
# The top ones that are high in sales are Furnishings and Fasteners.Their Profits are also above average mean line.
# The one which are both backward in Sales and profit are Labels ,Phones,Paper,Machines etc..
```

```python
# shipmode vs segment frequencies

contingency_table = pd.crosstab(ssdf['Ship Mode'], ssdf['Segment'])
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='flare_r')
plt.title('Ship Mode vs. Segment Frequencies')
plt.show()

# RESULT Analysis:
#This shows the heatmap of frequency of each combination of Ship Mode and Segment
# We can figure which shipping methods are most prefered and popular for each customer segment.
```

```python
#Sales vs Profit by Category and Discount

sns.scatterplot(data=ssdf, x='Sales', y='Profit', hue='Category', style='Segment')
plt.title('Sales vs. Profit by Category and Segment')
plt.show()

# RESULT Analysis:
#We can see that the Furniture does not show vast sales nor Profit regardless of Segment
#Most sales and profit in consumer segment is the technology category
```

```python
#pairplot

sns.pairplot(ssdf[['Sales', 'Profit', 'Quantity', 'Discount']])
plt.show()

# The pairplot shows all pairwise relationships between numerical variables, helping identify correlations and patterns.
# It also highlights category differences, revealing how each group behaves across the variables.
```
