## **Outlier Removal**

Outlier Detection on Numerical

```python
import numpy as np
from scipy.stats import median_abs_deviation
from scipy.stats.mstats import winsorize


def iqr_clipper(arr):
  q1 = np.quantile(arr, 0.25)
  q3 = np.quantile(arr, 0.75)
  iqr = q3 - q1
  lowerbound = q1 - (1.5 * iqr)
  upperbound = q1 + (1.5 * iqr)
  return np.clip(arr, lowerbound, upperbound)

def zscore_clipper(arr):
  arr_mean = np.mean(arr)
  arr_std = np.std(arr)
  lowerbound = arr_mean - (3.5 * arr_std)
  upperbound = arr_mean + (3.5 * arr_std)
  return np.clip(arr, lowerbound, upperbound)

cols_to_iqr_clip = ['No of Bedrooms', 'No of Bathrooms', 'Overall Grade']
cols_to_zscore_clip = ['Latitude', 'Longitude']
```

```python
for col in cols_to_iqr_clip:
  housePrice_df[col] = iqr_clipper(housePrice_df[col])

for col in cols_to_zscore_clip:
  housePrice_df[col] = zscore_clipper(housePrice_df[col])
```

```python
housePrice_df['Living Area after Renovation (in Sqft)'] = winsorize(housePrice_df['Living Area after Renovation (in Sqft)'], limits = (0.05, 0.05))
housePrice_df['Lot Area after Renovation (in Sqft)'] = winsorize(housePrice_df['Lot Area after Renovation (in Sqft)'], limits = (0.05, 0.1))
housePrice_df['Renovated Area of the House from Basement (in Sqft)'] = winsorize(housePrice_df['Renovated Area of the House from Basement (in Sqft)'], limits = (0.05, 0.05))
```
